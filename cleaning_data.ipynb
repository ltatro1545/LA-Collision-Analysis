```python
import pandas as pd
import numpy as np
```

```python
# Import csv file from Resource folder to create DataFrame
file_loc = "Resources/Parties.csv"
parties_df = pd.read_csv(file_loc)
parties_df
```

```python
# Replace "-" with NaN
parties_df = parties_df.replace(r'-', np.NaN, regex=True)
parties_df
```

```python
# Inspect the data types in the parties_df DataFrame
parties_df.dtypes
```

```python
# Determine the number of null values in the dataset
parties_df.isnull().sum()
```

```python
# Determine the number of null values in the dataset
parties_df.isnull().sum()
```

```python
# Drop NaN values
parties_df.dropna(inplace=True)
parties_df
```

```python
# Change ACCIDENT_YEAR to datetime
parties_df["ACCIDENT_YEAR"] = pd.to_datetime(parties_df["ACCIDENT_YEAR"], format='%Y')
parties_df
```

```python
# Get list of column names and data types
parties_df.dtypes
```

```python
# Import csv file from Resource folder to create DataFrame
file_loc = "Resources/Crashes_2017.csv"
crashes_2017_df = pd.read_csv(file_loc)
crashes_2017_df
# Import csv file from Resource folder to create DataFrame
file_loc = "Resources/Crashes_2018.csv"
crashes_2018_df = pd.read_csv(file_loc)
crashes_2018_df
# Import csv file from Resource folder to create DataFrame
file_loc = "Resources/Crashes_2019.csv"
crashes_2019_df = pd.read_csv(file_loc)
crashes_2019_df
```

```python
crashes_df = pd.concat([crashes_2017_df, crashes_2018_df, crashes_2019_df])
crashes_df
```

```python
# Replace "-" with NaN
crashes_df = crashes_df.replace(r'-', np.NaN, regex=True)
crashes_df
```

```python
# Determine the number of null values in the dataset
crashes_df["PCF_CODE_OF_VIOL"].isnull().sum()
```

```python
crashes_df = crashes_df.drop(columns=["PROC_DATE", "COLLISION_DATE", 
                                      "LONGITUDE", "LATITUDE", "SECONDARY_RAMP", 
                                      "PRIMARY_RAMP", "ROAD_COND_2", "SPECIAL_COND",
                                     "BEAT_TYPE", "CHP_BEAT_TYPE", "CHP_BEAT_CLASS",
                                     "WEATHER_2", "STATE_HWY_IND","CALTRANS_COUNTY", 
                                     "CALTRANS_DISTRICT", "STATE_ROUTE", "ROUTE_SUFFIX",
                                     "POSTMILE_PREFIX", "POSTMILE", "LOCATION_TYPE",
                                     "RAMP_INTERSECTION", "SIDE_OF_HWY", "PCF_CODE_OF_VIOL",
                                     "CHP_ROAD_TYPE", "DIRECTION", "CITY_DIVISION_LAPD",
                                     "BEAT_NUMBER", "REPORTING_DISTRICT", "PCF_VIOL_SUBSECTION",
                                     "PRIMARY_RD"])
crashes_df  
```

```python
# Determine the number of null values in the dataset
crashes_df.isnull().sum()
```

```python
crashes_df["PEDESTRIAN_ACCIDENT"].fillna(0, inplace=True)
crashes_df["PEDESTRIAN_ACCIDENT"].replace("Y", 1,inplace=True)
crashes_df["BICYCLE_ACCIDENT"].fillna(0, inplace=True)
crashes_df["BICYCLE_ACCIDENT"].replace("Y", 1,inplace=True)
crashes_df["MOTORCYCLE_ACCIDENT"].fillna(0, inplace=True)
crashes_df["MOTORCYCLE_ACCIDENT"].replace("Y", 1,inplace=True)
crashes_df["TRUCK_ACCIDENT"].fillna(0, inplace=True)
crashes_df["TRUCK_ACCIDENT"].replace("Y", 1,inplace=True)
crashes_df["ALCOHOL_INVOLVED"].fillna(0, inplace=True)
crashes_df["ALCOHOL_INVOLVED"].replace("Y", 1,inplace=True)
crashes_df["STWD_VEHTYPE_AT_FAULT"].fillna(0, inplace=True)
crashes_df["CHP_VEHTYPE_AT_FAULT"].fillna(0, inplace=True)
```

```python
# Determine the number of null values in the dataset
crashes_df.isnull().sum()
```
